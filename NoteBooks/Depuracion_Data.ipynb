```python
import pandas as pd
import numpy as np
```

```python
# Extraer los datros del archivo CS
data  = pd.read_csv("../data/raw/WB_HCP_HYPERTENSION.csv")
```

```python
clean_columns = [
    "REF_AREA_LABEL",
    "TIME_PERIOD",
    "SEX_LABEL",
    "AGE_LABEL",
    "OBS_VALUE",
    "UNIT_MEASURE_LABEL"
]

df = data[clean_columns].copy()
```

```python
df.rename(columns={
    "REF_AREA_LABEL": "country",
    "TIME_PERIOD": "year",
    "SEX_LABEL": "sex",
    "AGE_LABEL": "age_group",
    "OBS_VALUE": "hypertension_prevalence",
    "UNIT_MEASURE_LABEL": "unit"
}, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df["sex"].value_counts()
```

```python
sex_mapping = {
    "Total": "Both sexes",
    "Both sexes": "Both sexes",
    "Male": "Male",
    "Female": "Female"
}
df["sex"] = df["sex"].map(sex_mapping)
```

```python
df["sex"].value_counts(dropna=False)
```

```python
df_total = df[df["sex"] == "Both sexes"].copy()
```
