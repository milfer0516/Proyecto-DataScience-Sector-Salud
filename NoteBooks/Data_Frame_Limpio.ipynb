```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/hypertension_clean_1990_2019.csv")
```

```python
features = (
    df
    .groupby("country")["hypertension_prevalence"]
    .agg(
        mean_prevalence="mean",
        std_prevalence="std",
        min_prevalence="min",
        max_prevalence="max"
    )
    .reset_index()
)
```
